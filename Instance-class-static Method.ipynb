# Instance Method


```python
class student():
        
    school = 'nit'   
    
    def __init__(self, m1, m2, m3): 
        self.m1 = m1
        self.m2 = m2
        self.m3 = m3

    def avg(self):   
        return (self.m1+self.m2+self.m3)/3
    
s1 = student(10, 20, 30)
s2 = student(5, 15, 10)

print(s1.avg())
print(s2.avg())

```

# class method

```python
class student(): 
        
    school = 'nit'
    
    def __init__(self, m1, m2, m3) : # hear i pass 3 variable 
        self.m1 = m1
        self.m2 = m2
        self.m3 = m3
#lets create a method to calculate average of the marks 
    def avg(self):   # ------> hear this average is INSTANCE METHOD 
        return (self.m1+self.m2+self.m3)/3
    
    def info(cls): # user need to define cls - for class variable 
        return cls.school 
        
# lets create  object hear as we created s1, s2 we created 2 object & also i want to pass value in this  
s1 = student(10, 20, 30)
s2 = student(5, 15, 10)

print(s1.avg())
print(s2.avg())

print(s1.info())
print(s2.info())
```

```python
class student(): 
        
    school = 'nit'
    
    def __init__(self, m1, m2, m3) : # hear i pass 3 variable 
        self.m1 = m1
        self.m2 = m2
        self.m3 = m3
#lets create a method to calculate average of the marks 
    def avg(self):   # ------> hear this average is INSTANCE METHOD 
        return (self.m1+self.m2+self.m3)/3
    
    @classmethod   #Decoretor = you can use decoretor in class method and static method not for instance method 
    def info(cls): # user need to define cls - for class variable 
        return cls.school 
        
# lets create  object hear as we created s1, s2 we created 2 object & also i want to pass value in this  
s1 = student(10, 20, 30)
s2 = student(5, 15, 10)


print(s1.info())
print(s1.avg())
```

# Static method


```python
class student(): 
        
    school = 'nit'
    
    def __init__(self, m1, m2, m3) : # hear i pass 3 variable 
        self.m1 = m1
        self.m2 = m2
        self.m3 = m3
#lets create a method to calculate average of the marks 
    def avg(self):   # ------> hear this average is INSTANCE METHOD 
        return (self.m1+self.m2+self.m3)/3
    
    @classmethod   #Decoretor = you can use decoretor in class method and static method not for instance method 
    def info(cls): # user need to define cls - for class variable 
        return cls.school 
        
    @staticmethod
    def stat():
        print('this is static method')
        
# lets create  object hear as we created s1, s2 we created 2 object & also i want to pass value in this  
s1 = student(10, 20, 30)
s2 = student(5, 15, 10)

print(student.stat())
print(s1.info())
print(s1.avg())
```
