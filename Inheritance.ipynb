
# 1. Single Level Inheritance 


```python
class A():
    def feature1(self):
        print('Feature1 is hi')
    def feature2(self):
        print('Featuer2 is hi')


class B(A):    # class B inheritace of class A
    def feature3(self):
        print('Feature3 is hello')
    def feature4(self):
        print('Featuer4 is hello')

a1 = A()
b1 = B() 


a1.feature1()
a1.feature2()

b1.feature3()
b1.feature4()
b1.feature2()
b1.feature1()
```

# Malti level inheritace

```python
class A():
    def feature1(self):
        print('Feature1 is hi')
    def feature2(self):
        print('Featuer2 is hi')


class B():    
    def feature3(self):
        print('Feature3 is hello')
    def feature4(self):
        print('Featuer4 is hello')

class C(B):   # class C inheritace of class B
    def feature5(self):
        print('Feature5 is bye')


a1 = A()
b1 = B() 
c1 = C()

a1.feature1()
a1.feature2()

b1.feature3()
b1.feature4()

c1.feature5()
c1.feature4()
c1.feature3()


```

# 3. Multiple Inheritance 


```python
class A():
    def feature1(self):
        print('Feature1 is hi')
    def feature2(self):
        print('Featuer2 is hi')


class B():    
    def feature3(self):
        print('Feature3 is hello')
    def feature4(self):
        print('Featuer4 is hello')

class C(A,B):     #one class ,we can inherit more than two classes
    def feature5(self):
        print('Feature5 is bye')


a1 = A()
b1 = B() 
c1 = C()

#a1.feature1()
#a1.feature2()

#b1.feature3()
#b1.feature4()

print(c1.feature5())
print(c1.feature4())
print(c1.feature3())
print(c1.feature2())
print(c1.feature1())

```

# Super Function in inheritance

```python

class A():
    def __init__(self):
        print('in A init')
    def feature1(self):
        print('Feature1 is hi')
    def feature2(self):
        print('Featuer2 is hi')


class B(A):   
    def __init__(self):
       super().__init__()
       print('in B init')
    def feature3(self):
        print('Feature3 is hello')
    def feature4(self):
        print('Featuer4 is hello')


b1 = B() 



```

# Method Resolution Order (MRO) 

```python
class A():
    def __init__(self):
        print('in A init')
    def feature1(self):
        print('Feature1 is hi')
    def feature2(self):
        print('Featuer2 is hi')


class B():  
    def __init__(self):
       print('in B init')  
    def feature3(self):
        print('Feature3 is hello')
    def feature4(self):
        print('Featuer4 is hello')

class C(A,B):     
    def __init__(self):
       super().__init__()
       print('in C init')
    


a1 = C()

```
