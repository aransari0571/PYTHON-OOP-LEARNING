```python
class computer():
    def __init__(self):
        self.name = 'prakash'
        self.age = 40

c1 = computer()
c2 = computer()


print(c1.name, c1.age)
print(c2.name, c2.age)
```

```python
class computer():
    def __init__(self):
        self.name = 'prakash'
        self.age = 40

c1 = computer()
c2 = computer()

c1.name= 'senapati'
c1.age = 41

print(c1.name, c1.age)
print(c2.name, c2.age)
```

```python
class computer():
    def __init__(self):
        self.name = 'prakash'
        self.age = 40

c1 = computer()
c2 = computer()

self = computer()
self.name= 'senapati'
self.age = 41

print(c1.name, c1.age)

print(self.name, self.age)
```
