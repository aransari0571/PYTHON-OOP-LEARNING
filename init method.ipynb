```python
class computer():    
    
    def __init__(self, cpu, ram, price):
        print('in init')
        
    def config(self):     
        print('i5', 8)
        print('i7', 16)
        print('i7', 30000)
        
std1 = computer('i5', 8, 30000)
std2 = computer('i7', 16, 50000)

std1.config() 
#std2.config()
```

```python
class computer():        
    
    def __init__(self,cpu,ram, price): # cpu & ram as argument
        self.cpu = cpu
        self.ram = ram 
        self.price= price
        #print('init')
    
    def config(self):     
        print('config is', self.cpu, self.ram, self.price)

com1 = computer('i5', 8, 2000)
com2 = computer('i7', 16, 3000)

com1.config() 
com2.config()
```
